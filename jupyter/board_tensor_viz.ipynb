```python
import chess
import chess.svg
import matplotlib.pyplot as plt
from dinora.board_representation2 import board_to_compact_state, compact_state_to_board_tensor
board_to_tensor = lambda x: compact_state_to_board_tensor(board_to_compact_state(x))
```

```python
board = chess.Board('2rr1bk1/p2q1p2/1p4p1/3P3p/5N2/PQ2B2P/5PP1/3R2K1 b - - 2 27')
tensor = board_to_tensor(board)
chess.svg.board(board, size=350, flipped=True)
```

```python
images = tensor
fig, axes = plt.subplots(3, 6, figsize=(12, 6))
axes = axes.flatten()

for i in range(18):
    axes[i].imshow(images[i])

plt.show()
```
